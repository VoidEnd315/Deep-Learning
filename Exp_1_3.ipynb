## LINEAR REGRESSION

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5]).reshape(-1,1)
print(x)
y = np.array([2,4,5,4,5])
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
t = [[6],[7],[8],[2],[4]]
predictions = model.predict(t)
print(predictions)
```

```python
plt.scatter(t,predictions,label ='prediction',color = 'green')
prediction = model.predict(x)
plt.scatter(x,y,label = 'Actual Data', color = 'blue')
plt.plot(x,prediction,label='Regression Line',color = 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear regression')
plt.legend()
plt.show()
```

## K-MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
```

```python
data = [[1.2,2.3],[2.8,3.6],[3.5,4.7],[2.8,3.6],[5.1,6.0]]
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data)
```

```python
labels = kmeans.labels_
print(labels)
```

```python
cluster_centers = kmeans.cluster_centers_
```

```python
colors = ['b' if label == 0 else 'r' for label in labels]
plt.scatter(*zip(*data), c=colors, marker='o', label='Data Points')
plt.scatter(*zip(*cluster_centers), color='g', marker='x',s=1000,  label='Cluster Centers')
plt.title('K-Means Clustering')
```

# Exp. 2

## LINEAR REGRESSION with Synthetic Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.linspace(0,10,50)
y = 2*x+1+np.random.normal(0,1,50)
```

```python
def simple_linear_regression(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    numerator = np.sum((x-x_mean)*(y-y_mean))
    denominator = np.sum((x-x_mean)**2)

    slope = numerator/denominator
    intercept = y_mean-slope*x_mean

    return slope,intercept

slope,intercept = simple_linear_regression(x,y)
```

```python
predictions = slope*x+intercept
```

```python
plt.scatter(x,y,label='Original Data')
plt.plot(x, predictions, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

## Linear Regression using sin(x)


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.linspace(0,10,50)
print(len(x))
y = np.sin(x)
```

```python
plt.scatter(x,y,label='Original Data')
plt.plot(x, predictions, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Set a random seed for reproducibility
np.random.seed(42)

# Generate some data
x = np.linspace(0, 10, 50).reshape(-1, 1)
y = np.sin(x)

# Initialize a dictionary to store models
models = {}

# Plot the original sin(x) data
plt.figure(figsize=(12, 8))
plt.scatter(x, y, label='sin(x)')

# Fit and plot polynomial regression models for various degrees
for degree in [1, 3, 5, 7, 10]:
    poly_features = PolynomialFeatures(degree=degree)
    x_poly = poly_features.fit_transform(x)

    model = LinearRegression()
    model.fit(x_poly, y)  # Corrected: 'fit' instead of 'fit_transform'
    y_poly_pred = model.predict(x_poly)

    plt.plot(x, y_poly_pred, label="Degree " + str(degree))
    models[degree] = model

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin function vs. polynomial Regression')
plt.show()
```

# Exp. 3 - Image Enhancement Techniques

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image_path = "/content/image.jpg"
image = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)

print(image.shape)
equalized_image = cv2.equalizeHist(image)

kernel = np.ones((5,5), np.uint8)


dilated_image = cv2.dilate(image,kernel,iterations=1)
eroded_image = cv2.erode(image,kernel,iterations=1)

plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.hist(image)
plt.title("Histogram")

plt.subplot(2,3,2)
plt.imshow(image,cmap="gray")
plt.title("Original image")

plt.subplot(2,3,3)
plt.imshow(equalized_image,cmap="gray")
plt.title("Equalized image")

plt.subplot(2,3,4)
plt.imshow(dilated_image,cmap="gray")
plt.title("Dilated image")

plt.subplot(2,3,5)
plt.imshow(eroded_image,cmap="gray")
plt.title("Eroded image")

plt.tight_layout()
plt.show()
```
